### 데이터 청크 수집 

```python
poem = []
chunksize = 5
fin = open("/content/테스트용(txt_신한카드내국인).txt",'rt', encoding='cp949')
chunk = 1
for x in range(chunksize):
  line = fin.readline()
  if not line:
    break
  poem.append(line) 
fin.close()

headers = pd.DataFrame(poem)[0].apply(lambda x: x.split('\t')).iloc[0]
first_value = pd.DataFrame(poem)[0].apply(lambda x: x.split('\t')).iloc[1]

hi = pd.DataFrame(first_value,headers).T
for i in range(2,chunksize):
  hi.loc[i-1,:] = pd.DataFrame(poem)[0].apply(lambda x: x.split('\t')).iloc[i]
hi
```
